# Data Visualization in Jupyter Notebook

## Bar Graph

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60]
y = [13, 45, 23, 34, 96, 76]
plt.title('Bar Graph')
plt.bar(x, y, color='dodgerblue', width=5)
plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt
x = [35, 20, 30, 40, 50, 30]
y = ['Apple', 'Bananna', 'Grapes', 'Orange', 'PineApple', 'Dragon Fruit']
plt.title('Pie Chart')
plt.pie(x, labels=y)
plt.show()
```

## Line Chart

```python
import matplotlib.pyplot as plt
values = [35, 20, 30, 25, 50, 30]
label = ['Apple', 'Bananna', 'Grapes', 'Orange', 'PineApple', 'Dragon Fruit']
plt.title('Line Chart')
plt.plot(label,values,color='hotpink')
plt.show()
```

## Scatter Graph

```python
import matplotlib.pyplot as plt
values = [35, 20, 30, 25, 15, 30]
label = ['Apple', 'Bananna', 'Grapes', 'Orange', 'PineApple', 'Dragon Fruit']
plt.title('Scatter Graph')
plt.scatter(label,values)
plt.show()
```
